# 74. Search a 2D Matrix
You are given an m x n integer matrix matrix with the following two properties:

Each row is sorted in non-decreasing order.
The first integer of each row is greater than the last integer of the previous row.
Given an integer target, return true if target is in matrix or false otherwise.

You must write a solution in O(log(m * n)) time complexity.

```python
# 74. Search a 2D Matrix

class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:

        left = 0
        right = len(matrix)
        while left < right:
            middle = left + (right - left) // 2
            middle_arr = matrix[middle]
            if middle_arr[0] <= target <= middle_arr[-1]:
                if target in set(middle_arr):
                    return True
                else:
                    return False
 
            elif middle_arr[0] > target:
                right = middle 
            elif middle_arr[-1] < target:
                left = middle + 1
  # time O(log(n)), space O(1)
```

# 189. Rotate Array

To rotate the array in place, I create the reverse method and use this to reverse the array at key points plus use % modulo to reduce 
unnecessary rotations. 

```python
# 189. Rotate Array
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        nums_len = len(nums)
        k %= nums_len

        self.reverse(nums, 0, nums_len - 1)
        self.reverse(nums, 0, k - 1)
        self.reverse(nums, k, nums_len - 1)


    def reverse(self, array: list, start: int, end: int) -> None:
        while start < end:
            array[start], array[end] = array[end], array[start]
            start += 1
            end -= 1
            
# time O(n) , space O(1)
# runtime: 359ms, beats 61%
# memory: 25MB, beats 78% 
```

"a" if a == b

```python
# 163. Missing Ranges

class Solution:
    def findMissingRanges(self, nums: List[int], lower: int, upper: int) -> List[str]:
        output = []
        counter = lower
        for number in nums:
            if number != counter:
                oneLess = number - 1
                if oneLess == counter:
                    output.append(str(oneLess))
                else:
                    output.append(str(counter) + "->" + str(oneLess))
                counter = number + 1            
            else:
                counter += 1
        if counter < upper:
            output.append(str(counter) + "->" +str(upper))
        elif counter == upper:
            output.append(str(upper))
        return output
      
# Time O(N), Space O(1)
```

# 11. Container With Most Water
You are given an integer array height of length n. There are n vertical lines drawn such that the two endpoints of the ith line are (i, 0) and (i, height[i]).

Find two lines that together with the x-axis form a container, such that the container contains the most water.

Return the maximum amount of water a container can store.

Notice that you may not slant the container.

```python
# 11. Container With Most Water

class Solution:
  def maxArea(self, height: List[int]) -> int:
    left = 0
    right = len(height) - 1
    water = 0
    while left < right:
      water = max(min(height[left], height[right]) * (right - left), water)
      if height[left] < height[right]:
        left += 1
      else:
        right -= 1
    return water
  
# Time O(N), Space O(1)
```

# 540. Single Element in a Sorted Array

You are given a sorted array consisting of only integers where every element appears exactly twice, except for one element which appears exactly once.

Return the single element that appears only once.

Your solution must run in O(log n) time and O(1) space.

```python
# 540. Single Element in a Sorted Array

class Solution:
    def singleNonDuplicate(self, nums: List[int]) -> int:
        l = 0
        r = len(nums) - 1
        while l < r:
            m = (l + r) // 2
            if m % 2 == 1:
                m -= 1
            if nums[m] == nums[m + 1]:
                l = m + 2
            else:
                r = m
        return nums[l]

# Time O(n), Space O(1)
```

An island is surrounded by water and is formed by connecting adjacent lands horizontally or vertically. You may assume all four edges of the grid are all surrounded by water.

```python
# 200. Number of Islands

class Solution:
    def numIslands(self, grid: List[List[str]]) -> int:
        cnt = 0
        r = len(grid)
        c = len(grid[0])

        def BFS(row, col):
            up = [-1,0,1,0]
            dn = [0,1,0,-1]
            queue = [[row,col]]
            while queue != []:
                node = queue.pop(0)
                nodeRow = node[0]
                nodeCol = node[1]

                for i in range(len(up)):
                    mRow = nodeRow + up[i]
                    mCol = nodeCol + dn[i]
                    if 0 <= mRow < r and 0 <= mCol < c:
                        if grid[mRow][mCol] == "1":
                            grid[mRow][mCol] = "0"
                            queue.append([mRow,mCol])
            return
        
        for row in range(r):
            for col in range(c):
                if grid[row][col] == "1":
                    grid[row][col] = "0"
                    cnt += 1
                    BFS(row,col)

        return cnt
    
    
# Time: O(N*M), Space: O(1)
```

# 24. Swap Nodes in Pairs
Given a linked list, swap every two adjacent nodes and return its head.
You must solve the problem without modifying the values in the list's nodes (i.e., only nodes themselves may be changed.)


```python
# 24. Swap Nodes in Pairs

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def swapPairs(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if head is None or head.next is None:
            return head
        else:
            hh = head.next
            prev = None
            while head is not None:
                if head.next is not None:

                    tmp = head.next.next
                    tmp2 = head.next

                    tmp2.next = head
                    head.next = None

                    if prev is not None:
                        prev.next = tmp2

                    prev = head
                    head = tmp
                else:
                    prev.next = head
                    break
            return hh
          
# Time O(N), Space O(1)
```

# 129. Sum Root to Leaf Numbers
You are given the root of a binary tree containing digits from 0 to 9 only.

Each root-to-leaf path in the tree represents a number.

For example, the root-to-leaf path 1 -> 2 -> 3 represents the number 123.
Return the total sum of all root-to-leaf numbers. Test cases are generated so that the answer will fit in a 32-bit integer.

A leaf node is a node with no children.

```python
# 129. Sum Root to Leaf Numbers

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def sumNumbers(self, root: Optional[TreeNode]) -> int:
        ttl = [0]

        def helper(node, arr = [], ttl = 0):
            if node is not None:
                arr.append(str(node.val))
                if node.left is None and node.right is None:
                    tmpStr = "".join(arr)
                    tmpInt = int(tmpStr)
                    ttl += tmpInt
                    arr.pop()
                    return ttl
                else:
                    if node.left is not None:
                        ttl += helper(node.left, arr, 0)
                    if node.right is not None:
                        ttl += helper(node.right, arr, 0)
                    arr.pop()
                    return ttl
        return helper(root)
      
# Time O(N), Space O(N)
```

# 1041. Robot bounded in circle

```python
class Solution:
    def isRobotBounded(self, instructions: str) -> bool:
        # idx 0 = north, 1 = east, 2 = south, 3 = west

        # possible directions facing north, east, south, west
        direction = [[0,1], [1,0], [0,-1],[-1,0]]
        # imagine x,y graph starting at 0,0 coordinate
        x = y = 0
        # facing north
        idx = 0

        for i in instructions:
            if i == 'L': # update direction
                idx = (idx + 3) % 4
            elif i == 'R': # update direction
                idx = (idx + 1) % 4
            else:
                # update coordinate based on direction
                x += direction[idx][0]
                y += direction[idx][1]
        
        return (x == 0 and y == 0) or idx != 0
      
```
